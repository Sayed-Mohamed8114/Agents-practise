```python
# first download it from terminal : sudo apt install python3-tk
# check the version : python3 -m tkinter
# first we will import tkinter 
import tkinter as tk
```

```python
# first we make the main window to the app 
root = tk.Tk() # -> with this we make a window any thing will be added in it now like (buttons , Labels , Entry... )
```

```python
# tk.label()
tk.Label(root , text="enter pass: ").pack()
# label is a text that appear on the screen
# root is the parent of text
# Pack here is a way to arrange the item there are also (.grid(),.place())
```

```python
# to take an input from the user
entry = tk.Entry(root,show="*")
# Entry here is a place to input in it  
# show ="*" is to display the input like ** as the real passwords
# after the input we will use entry.get to read the text from the user -> will be in the check_password function 
```

```python
# Buttons
tk.Button(root, text="Check", command="").pack()
# text= "check" is the text that will appear on the button , command -> is the function that will be executed when we will click on the button
```

```python
# widget.get()
pwd = entry.get()
# is the way we will take the text from the user 
# every entry has its own get
```
